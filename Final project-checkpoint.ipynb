```python
def dfAlone (temp):
    temp= temp.iloc[700:]   #Start after the first 7 seconds
    temp= temp.reset_index(drop=True)
    temp= temp.drop([' Frame ID'],axis=1)
    RightHandTemp= RightHand.iloc[:len(temp)-1]  
    RightHandTemp= RightHandTemp.reset_index(drop=True)
    RightHandTemp= RightHandTemp.drop([' Frame ID'],axis=1)
    temp= pd.merge(temp, RightHandTemp, left_index=True, right_index=True)  

    return temp
```

```python
def dfTwoHands (temp):
    temp= temp.iloc[700:]   #Start after the first 7 seconds
    dfRight= temp[temp[" Hand Type"]=="right"]
    dfLeft= temp[temp[" Hand Type"]=="left"]
    dfRight.set_index(' Frame ID',inplace=True)
    dfLeft.set_index(' Frame ID',inplace=True)
    temp= pd.merge(dfLeft, dfRight, how='left',on=' Frame ID') 
    
    return temp
```

```python
State= ['Alone', 'Spontan', 'Sync']
```

```python
#Delete the '# hands' columns, to avoid overfitting.
train= train.drop([' # hands_x', ' # hands_y'], axis=1)


#Delete the 'Time' columns, the time isn't important for the predictions
train= train.drop(['Time_x', 'Time_y'], axis=1)
```

```python
#Changing the 'Hand Type' columns to a float data type

train[' Hand Type_x'] = train[' Hand Type_x'].map({     
        'left': 1
})

train[' Hand Type_y'] = train[' Hand Type_y'].map({     
        'right': 2
})


#Changing the 'State' columns to a float data type
train['State']= pd.to_numeric(train['State'])
```

```python
train.info()
```

```python
#Now that we have only the columns that are necessary for the prediction, we can't have any missing values
train= train.dropna()
train.info()
```

```python
#Because we have 100 samples for every second, we don't need all of the samples 

Train= train[::10]    
```

```python
Train
```

```python
print('Final train data -', Train.shape)
```

```python
#Creating the valitation df

test= pd.concat([df, TestAlone, TestSpontan, TestSync], sort=False, ignore_index=True)

test= test.drop([' # hands_x', ' # hands_y'], axis=1)
test= test.drop(['Time_x', 'Time_y'], axis=1)

test[' Hand Type_x'] = test[' Hand Type_x'].map({     
        'left': 1
})

test[' Hand Type_y'] = test[' Hand Type_y'].map({     
        'right': 2
})

test['State']= pd.to_numeric(test['State'])

test= test.dropna()

Test= test[::10]       

print('Final test data -', Test.shape)
```

```python
#Prepare the training and testing dataset 
X_train = Train.drop(columns='State')
y_train = Train['State']

X_test = Test.drop(columns='State')
y_test = Test['State']
```

## Making Predictions

```python
#NAÏVE BAYES

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
nb_predictions= nb.predict(X_test)
```

```python
print(classification_report(y_test, nb_predictions, target_names= State))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, nb_predictions)
```

```python
print(confusion_matrix(y_test, nb_predictions))
```

```python
from sklearn.model_selection import cross_val_score

nb_scores= cross_val_score(nb, X_train, y_train, cv=3, scoring="accuracy")
print(nb_scores)
print('mean- ',nb_scores.mean())
```

```python
#ADABOOST

from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(X_train, y_train)
ada_predictions= ada.predict(X_test)
```

```python
print(classification_report(y_test, ada_predictions, target_names= State))
```

```python
accuracy_score(y_test, ada_predictions)
```

```python
print(confusion_matrix(y_test, ada_predictions))
```

```python
ada_scores= cross_val_score(ada, X_train, y_train, cv=3, scoring="accuracy")
print(ada_scores)
print('mean- ',ada_scores.mean())
```

```python
#RANDOM FOREST

#Creating the model with 100 trees
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
rf_predictions= rf.predict(X_test)
```

```python
print(classification_report(y_test, rf_predictions, target_names= State))
```

```python
accuracy_score(y_test, rf_predictions)
```

```python
print(confusion_matrix(y_test, rf_predictions))
```

```python
rf_scores= cross_val_score(rf, X_train, y_train, cv=3, scoring="accuracy")
print(rf_scores)
print('mean- ',rf_scores.mean())
```

```python
#KNN      

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
knn_predictions= knn.predict(X_test)
```

```python
print(classification_report(y_test, knn_predictions, target_names= State))
```

```python
accuracy_score(y_test, knn_predictions)
```

```python
print(confusion_matrix(y_test, knn_predictions))
```

```python
knn_scores= cross_val_score(knn, X_train, y_train, cv=3, scoring="accuracy")
print(knn_scores)
print('mean- ',knn_scores.mean())
```

```python
#PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
#NAÏVE BAYES- PCA

nb2 = GaussianNB()
nb2.fit(X_train_reduced, y_train)
nb2_predictions= nb2.predict(X_test_reduced)
accuracy_score(y_test, nb2_predictions)
```

```python
#Naive bayes works better without PCA
```

```python
#ADABOOST- PCA

ada2 = AdaBoostClassifier(n_estimators=100)
ada2.fit(X_train_reduced, y_train)
ada2_predictions= ada2.predict(X_test_reduced)
accuracy_score(y_test, ada2_predictions)
```

```python
#Adaboost works better when using PCA, we now have 0.718 accuracy, compared to 0.678 accuracy without PCA.
```

```python
#RANDOM FOREST- PCA

rf2 = RandomForestClassifier(n_estimators=100)
rf2.fit(X_train_reduced, y_train)
rf2_predictions= rf2.predict(X_test_reduced)
accuracy_score(y_test, rf2_predictions)
```

```python
#Random Forest works a lot better without PCA
```

```python
#KNN- PCA      

knn2 = KNeighborsClassifier(n_neighbors=3)
knn2.fit(X_train_reduced, y_train)
knn2_predictions= knn2.predict(X_test_reduced)
accuracy_score(y_test, knn2_predictions)
```

```python
#KNN works better without PCA
```
